## Question 1. Computing Transition in a Representative Agent Economy
Consider the following closed optimal growth economy populated by a large number of identical infinitely lived households that maximize: 

\begin{align*}
    E_0\bigg\{\sum_{t=0}^\infty \beta^tu(c_t)\bigg\}
\end{align*}

over consumption and leisure $u(c_t)=ln c_t$, subject to:

\begin{align*}
    c_t + i_t  = y_t
\end{align*}

\begin{align*}
    y_t= k_t^{1-\theta}(z h_t)^\theta
\end{align*}

\begin{align*}
    i_t= k_{t+1} - (1-\delta) k_t
\end{align*}

Set labor share to $\theta=0.67$. Also, to start with, set $h_t=0.31$ for all $t$. Population does not grow.

### (a) Compute the steady-state. Choose $z$ to match an annual capital-output ratio of 4, and an investment-output ratio of 0.25

The problem is: 

\begin{align*}
    \max_{c_t,k_{t+1}}  E_0\bigg\{\sum_{t=0}^\infty \beta^tu(c_t)\bigg\}
\end{align*}

Subject to:

\begin{align*}
    c_t + k_{t+1} = k_{t}^{1-\theta} (z h_t)^\theta + (1-\delta)k_t
\end{align*}

Notice that since the constraint holds with equality, we can isolate consumption and introduce it in the objective function. So we have:

\begin{align*}
     c_t  = k_{t}^{1-\theta} (z h_t)^\theta + (1-\delta)k_t-k_{t+1}
\end{align*}

And the problem becomes:

\begin{align*}
    \max_{k_{t+1}}  E_0\bigg\{\sum_{t=0}^\infty \beta^tu(k_{t}^{1-\theta} (z h_t)^\theta + (1-\delta)k_t-k_{t+1})\bigg\}
\end{align*}

The first order condition gives:

\begin{align*}
   \frac{\partial L}{\partial k_{t+1}}=0 \Longleftrightarrow
\end{align*}

\begin{align*}
   -\beta^t u'(k_{t}^{1-\theta} (z h_t)^\theta + (1-\delta)k_t-k_{t+1}) + \beta^{t+1}u'(k_{t+1}^{1-\theta} (z h_{t+1})^\theta + (1-\delta)k_{t+1}-k_{t+2})((1-\theta)k_{t+1}^{-\theta}(z h_{t+1})^{\theta}+(1-\delta))=0
\end{align*}

Which is in fact the Euler equation and can be written as: 

\begin{align*}
     \beta u'(c_{t+1})((1-\theta)k_{t+1}^{-\theta}(z h_{t+1})^{\theta}+(1-\delta))= u'(c_{t})
\end{align*}

And now we can impose steady state, which imply that $c_t=c_{t+1}=c^*$ and $k_{t}=k_{t+1}=k^*$, we get:

\begin{align*}
     \beta u'(c^*)((1-\theta)k^{*-\theta}(z h^*)^{\theta}+(1-\delta))= u'(c^*)
\end{align*}

\begin{align*}
    k^*=zh^*\bigg[\frac{(1-\theta)\beta}{1-(1-\delta)\beta}\bigg]^{\frac{1}{\theta}}
\end{align*}

Now that we have characterized the steady state level of capital. Now we  can impose that the capital-output ratio should be 4 and the investment-output 0.25. To do so, let's first normalize output equal to 1. By doing so, we get that $k=4$ and $i=0.25$. Now we can solve for the other parameters of the model: 

\begin{align*}
    i^* = \delta k^*  \leftrightarrow{} \delta  = \frac{0.25}{4}=0.0625
\end{align*}

\begin{align*}
    c^*= y^*-i^* = 1-0.25=0.75
\end{align*}

\begin{align*}
    y^*= k^{*1-\theta} (zh^*)^{\theta} \leftrightarrow{} z=\left(\frac{y}{k^{*1-\theta} h^{*\theta}}\right)^{\frac{1}{\theta}} = \left(\frac{1}{4^{1-0.67}0.31^{0.67}} \right)=1.629
\end{align*}

\begin{align*}
    \beta= \frac{1}{((1-\theta)k^{*-\theta}(zh^*)^{\theta})+(1-\delta)} = 0.98
\end{align*}

```python
# Define the parameters and values that I have computed analytically: 

θ=0.67       # Labor Share                                     
h=0.31       # Labor
k_i=4        # Capital at initial steady state
i_i=0.25     # Investment at initial steady state
c_i=0.75     # Consumption at initial steady state
y_i=1        # Normalized Output at intial steady state
δ=i_i/k_i        # Depreciation
z=1/((k_i**(1 - θ) * h ** θ) ** (1/θ))             # Labor productivity
β= 1 / ((1-θ) * k_i **(-θ) * (z*h) ** θ + 1 - δ)   # Discount factor

#Create a dictionary:

initia_state = {"labor share":θ, "capital":k_i, "labor":h, "investment":i_i, "consumption":c_i, "output":y_i, "depreciation":δ, "discount": β, "z":z}
```

So if we assume that the new $z=2z$ we get that the new steady state level of capital is:

\begin{align*}
    k_{new}^*=2zh^*\bigg[\frac{(1-\theta)\beta}{1-(1-\delta)\beta}\bigg]^{\frac{1}{\theta}}
\end{align*}

Which imply that $k^*$ has doubled and now it is $k^*=8$. Furthermore, we get that:

\begin{align*}
    y_{new}^* = (2k^*)^{1-\theta}(2h^*)^{\theta}=2(k^{*1-\theta} (zh^*)^{\theta})=2
\end{align*}

Since before it was normalized to 1 and now it has doubled. Also, regarding the new investment level we have:

\begin{align*}
   i_{new}^*= \delta k_{new}^* = 0.5
\end{align*}

And from this we get that the new consumption is: 

\begin{align*}
    c_{new}^*= y_{new}^*-i_{new}^*=1.5
\end{align*}

### (c) Compute the transition from the first to the second steady state and report the time-path for savings, consumption, labor and output.

To move from one steady state to another we will need to use the law of motion of capital. This law of motion needs to satisfy the optimality condition (Euler equation), so for this reason our transition path will be characterized by $n$ Euler equaitons, that will be written only as a function of capital. So take the Euler equation:

\begin{align*}
     \beta u'(c_{t+1})((1-\theta)k_{t+1}^{-\theta}(z h_{t+1})^{\theta}+(1-\delta))= u'(c_{t})
\end{align*}
And since the utility function is $u(c_t)= ln c_t$ we get that:

\begin{align*}
    c_{t+1}= \beta c_t((1-\theta)k_{t+1}^{-\theta}(z h_{t+1})^{\theta}+(1-\delta))
\end{align*}
Now we can substitute by consumption as a function of capital to get the Euler equation fully dependent on capital: 

\begin{align*}
    c_t  = k_{t}^{1-\theta} (z h_t)^\theta + (1-\delta)k_t-k_{t+1}
\end{align*}
So what we get is: 

\begin{align*}
   k_{t+1}^{1-\theta} (z h_{t+1})^\theta + (1-\delta)k_{t+1}-k_{t+2}= \beta (k_{t}^{1-\theta} (z h_t)^\theta + (1-\delta)k_t-k_{t+1})((1-\theta)k_{t+1}^{-\theta}(z h_{t+1})^{\theta}+(1-\delta))
\end{align*}

The procedure to characterize the transition path is the following:
1. We need to characterize $n$ different periods Euler equations, so will will have $n$ equations and $n+2$ unknowns
2. To have an identified system of equations, we can establish the initial capital level, which is the initial steady state and also the final capital level, which is the second steady state capital value.
3. Once we have a perfect identified system of $n$ equations and $n$ unknowns we can solve for it. Notice that our $n$ unknowns are nothing else than the transition values of capital towards the steady state.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
plt.style.use('ggplot')


#Characterize the transition path: 

n = 100 # Number of periods of transition

k = np.zeros(n) 

def transition_path(k,n=n):
    "This function returns the transition path towards the steady state over n periods"
    k[0] = k_i       # Initial Capital Level (First Steady State)
    k[n-1] = k_f     # Final Capital Level (Second Steady State)
    
    #Now let's generate the "n" different Euler equations: 
    trans = np.zeros(n)
    for i in range(n-2):
        if i == (n-2): #Since the other function is not well defined at n-2
            trans[i+1] = k[i+1]**(1-θ)*(z*h)**θ+(1-δ)*k[i+1]-k_f-β*(k[i]**(1-θ)*(z*h)**θ+(1-δ)*k[i]-k[i+1])*((1-θ)*k[i+1]**(-θ)*(z*h)**θ+1-δ)
        else:
            trans[i+1] = k[i+1]**(1-θ)*(z*h)**θ+(1-δ)*k[i+1]-k[i+2]-β*(k[i]**(1-θ)*(z*h)**θ+(1-δ)*k[i]-k[i+1])*((1-θ)*k[i+1]**(-θ)*(z*h)**θ+1-δ)
    return trans                        # trans is the system of equations that we need to solve

#Define an initial guess: 

x0 = np.linspace(4,8,n)                 # Grid of n points between4 and 8, which is the transition space. 
k_trans = fsolve(transition_path,x0)    # Solves the system of equations. Gives the transition path.


#Generate the transition path for the other variables:

## Output: 
y =np.zeros(n)
y[0]=y_i                                 # Initial Output level at First Steady State
for i in range(n -1):
    y[i+1]=k_trans[i]**(1-θ)*(z*h)**θ    # (y_{t} = k_{t}^(1-θ)(zh)^θ)

## Investment: 
i_path = np.zeros(n)
i_path[0] = i_i                                       # Initial Steady State level
for i in range (n-1):
    i_path[i+1] = k_trans[i+1] - (1-δ)*k_trans[i]     # (i_{t} = k_{t+1} - (1-δ)k_t)

## Consumption: 
c = np.zeros(n)
c[0] = c_i                             # Initial Steady State level
for i in range(n-1):
    c[i+1] = y[i+1] - i_path[i+1]      # (c_{t} = y_{t} - i_{t})


# Plots: 

fig, ax = plt.subplots()
ax.plot(k_trans,'m-o',label="Capital")
plt.ylabel("$k_t$",fontsize=15)
plt.xlabel("$t$",fontsize=15)
plt.legend()
plt.title("Capital Transition Path")


fig, ax = plt.subplots()
ax.plot(y,'r-o',label = "Output")
plt.ylabel("$y_t$",fontsize = 15)
plt.xlabel("$t$",fontsize = 15)
plt.legend()
plt.title("Output Transition Path")

fig, ax = plt.subplots()
ax.plot(i_path,'g-o',label = "Invesmtent")
plt.ylabel("$i_t$",fontsize = 15)
plt.xlabel("$t$",fontsize = 15)
plt.legend()
plt.title("Investment Transition Path")

fig, ax = plt.subplots()
ax.plot(c,'b-o',label = "Consumption")
plt.ylabel("$i_t$",fontsize = 15)
plt.xlabel("$t$",fontsize = 15)
plt.legend()
plt.title("Consumption Transition Path")


# More !! --> Show the overall effect on the economy:

# I will now consider that the economy was at the steady state for 10 peridos, and then introduce the shock:

i= 10       #Numer of periods at initial steady state

kss = np.ones(i)*4         # Capital Steady State
css = np.ones(i)*0.75      # Consumption Steady State
iss = np.ones(i)*0.25      # Investment Steady State
yss = np.ones(i)           # Output Steady State

k_extra = np.concatenate((kss,k_trans))
c_extra = np.concatenate((css,c))
i_extra = np.concatenate((iss,i_path))
y_extra = np.concatenate((yss,y))


trend_k = np.ones(n + i)*4 
trend_y = np.ones(n + i)*1 
trend_c = np.ones(n + i)*0.75 
trend_i = np.ones(n + i)*0.25 


fig = plt.figure(figsize=(8,6))
grid = np.arange(0, n+i, 1)
plt.plot(grid, k_extra, 'm', label='Capital')
plt.plot(grid, y_extra, color='r', label='Output')
plt.plot(grid, c_extra, color='b', label='Consumption')
plt.plot(grid, i_extra, color='g', label='Savings')
plt.plot(grid, trend_k, 'm--', linewidth=0.5)
plt.plot(grid, trend_y, 'r--',linewidth=0.5 )
plt.plot(grid, trend_c, 'b--',linewidth=0.5 )
plt.plot(grid, trend_i, 'g--',linewidth=0.5 )
plt.axvline(10, color='black')
plt.xlabel('$t$', fontsize =15)
plt.ylabel('$k_t$,$y_t$,$c_t$,$i_t$', fontsize=15)
plt.xlim(0, 105)
plt.legend(loc = 'upper right')
plt.title('Producitvity shock overall effect', fontsize=16)


plt.show()
```

## Question 2. Solve for the optimal COVID-19 lockdown model posed in the slides.

### (a) Show your results for a continuum of combinations of the $\beta \in [0,1]$ parameter (vertical axis) and the $c(TW) \in [0, 1]$ parameter (hztal axis). That is, plot for each pair of $\beta$ and $c(WT)$ the optimal allocations of $H, H_f , H_{nf} , H_f=H$, output, welfare, amount of infections and deaths. Note that if $H = N$ there is no lockdown, so pay attention to the potential non-binding constraint $H < N$. Discuss your results. Let's first present the model of the slides. The probelm of the social planner is :

\begin{align*}
    \max_{\{c,h_f,h_{nf}\}} \sum_i (c_i-k_fh_f-k_{nf}h_{nf}) -\omega D
\end{align*}
where $\omega$ denotes how much the planner cares about the number of deaths $D = D_{f}+ D_{nf}$.  Also we can use the aggregate resource constraint: 

\begin{align*}
    Y = \sum_i c_i = C
\end{align*}
And the labor market clearing condition is :

\begin{align*}
    \sum_{i \in j} h_i = H_{j} \quad \text{for} \quad j ={f,nf}
\end{align*}
So we get that the problem of the social planner is :

\begin{align*}
    \max_{\{H_{j}\in [0,N]\}_{j=\{f,nf\}}} \quad Y(H_{f},H_{nf}) - k_f H_f -k_{nf}H_{nf} -\omega D
\end{align*}
subject to:

\begin{align*}
    H = H_f +H_{nf} \leq N
\end{align*}
And now, we can use the fact that the production function has the form: 

\begin{align*}
    Y = \left(A_{f}H_{f}^{\frac{\rho-1}{\rho}} + c(TW) A_{f}H_{nf}^{\frac{\rho-1}{\rho}} \right)^{\frac{\rho }{\rho-1}}
\end{align*}
Where $c(TW) \in [0,1]$ is the productivity loss due to teleworking. Furthermore, we have that the probability of dying is: 

\begin{align*}
    D= (1-\gamma) I
\end{align*}
And I is the total number of infections that has the form: 

\begin{align*}
    I = iH_{f}
\end{align*}
Notice that $i$ is the unconditional infection rate at which a worker gets infected, and has the form: 
 
\begin{align*}
    i = \beta(HC)m(H_f)
\end{align*}
Where $m(H_{f})$ is the probability than an employed individual meets with a contagious individual at work and has the form:

\begin{align*}
    m(H_f)= \frac{i_0 H_{f}}{N}
\end{align*}
with $i_0$ being the initial share of infections at work. Also $\beta(HC) \in [0,1]$ is the conditional infection rate which depends on the extent of human contact. So after having considered all the previous equations we can rewrite the problem of the social planner as: 
\begin{align*}
    \max_{\{H_{j}\in [0,N]\}_{j=\{f,nf\}}} \quad \left(A_{f}H_{f}^{\frac{\rho-1}{\rho}} + c(TW) A_{f}H_{nf}^{\frac{\rho-1}{\rho}} \right)^{\frac{\rho}{\rho-1}} - k_f H_f -k_{nf}H_{nf} -\omega ((1-\gamma) \beta(HC)\frac{i_0 H_{f}^2}{N})
\end{align*}
subject to:

\begin{align*}
    H = H_f +H_{nf} \leq N
\end{align*}
So the Lagrangian of the problem is: 

\begin{align*}
    L = \left(A_{f}H_{f}^{\frac{\rho-1}{\rho}} + c(TW) A_{f}H_{nf}^{\frac{\rho-1}{\rho}} \right)^{\frac{\rho}{\rho-1}} - k_f H_f -k_{nf}H_{nf} -\omega \left((1-\gamma) \beta(HC)\frac{i_0 H_{f}^2}{N}\right) -\lambda [H_f +H_{nf} - N]
\end{align*}
The first order conditions give:
 \begin{align*}
     \frac{\partial L}{\partial H_f} =0 \leftrightarrow{} \left(A_{f}H_{f}^{\frac{\rho-1}{\rho}} + c(TW) A_{f}H_{nf}^{\frac{\rho-1}{\rho}} \right)^{\frac{-1}{\rho-1}} A_fH_f^{\frac{-1}{\rho}} -k_f -2\omega \left((1-\gamma) \beta(HC)\frac{i_0 H_{f}}{N}\right) -\lambda=0
 \end{align*}
  \begin{align*}
     \frac{\partial L}{\partial H_f} =0 \leftrightarrow{} \left(A_{f}H_{f}^{\frac{\rho-1}{\rho}} + c(TW) A_{f}H_{nf}^{\frac{\rho-1}{\rho}} \right)^{\frac{-1}{\rho-1}} c(TW)A_fH_{nf}^{\frac{-1}{\rho}} -k_{nf} -\lambda=0
 \end{align*}
 Together with the slackness conditions : 
 
\begin{align*}

\end{align*}
And now to solve it we just need to solve this system of equations.  Notice I will not do this analytically, since I have to solve for a continous range of values for two parameters. So I will use Python to compute the solution.

```python
# Question 2.

# Solve for β(HC) in [0,1] and c(TW) in [0,1]: 

from scipy.optimize import minimize      # Import the function to solve the problem

# Define a function to simplify:
# Define all the parameters following the exercice: 
Af=1
Anf=1
kf=0.2
knf=0.2
γ=0.9
i0=0.2
N=1


def heat(ω,ρ,n):
    "Produces heatmaps of the optimal values give omega,ro and the number of discrete points to evaluate"
    "Inputs--> Omega, Rho and n"
    "Output--> All heatmaps"
    # Generate the objective function: 
    
    func =lambda h:-1* ((Af*h[0]**((ρ-1)/ρ)+grid[j]*Anf*h[1]**((ρ-1)/ρ))**(ρ/(ρ-1)) - kf*h[0]-knf*h[1] - ω*((1-γ)*grid[i]*(i0*h[0]**2/N))) 

    # Generate the constraint:

    cons= ({'type':'ineq','fun': lambda h: N - h[0] - h[1] })

    # Discretize the continous domain of β(HC) and c(TW):

    grid = np.linspace(0,1,n)           # Discrete domain

    result1 = np.zeros(shape=(n,n))     # Array of results for hf

    result2 = np.zeros(shape=(n,n))     # Array of results for hnf

    for i in range(n):
            for j in range(n):
                h0 = [0.5,0.5]         # Initial guess
                bns = [(0,1),(0,1)]    # Bounds to the optimal values
                optimal = minimize(func,h0,constraints=cons,bounds=bns)      # Compute optimal hf and hnf 
                result1[i][j] = optimal.x[0]                      # Store the optimal hf
                result2[i][j] = optimal.x[1]                      # Store the optimal hnf
            
    # Now compute the other variables the question requires: 

    ## Output: 

    y_opt = np.zeros(shape=(n,n))   #Array for results of output

    for i in range (n):
        for j in range(n):
           y_opt[i][j]= (Af*result1[i][j]**((ρ-1)/ρ)+grid[j]*Anf*result2[i][j]**((ρ-1)/ρ))**(ρ/(ρ-1))

    ## H:

    H = result1 +result2
    
    ## Hf/H:

    Hf_H = result1/H

    ## Amount of infections: 

    I = np.zeros(shape=(n,n))    # Array for results of infections
    for i in range(n):
        for j in range(n):
            I[i][j] = result1[i][j]**2*i0*grid[i]

    ## Amount of deaths: 

    D  = (1-γ)*I

    ## Welfare:
    
    W = np.zeros(shape=(n,n))    # Array for results of Welfare
    
    for i in range(n):
        for j in range(n):
            W[i][j] = (Af*result1[i][j]**((ρ-1)/ρ)+grid[j]*Anf*result2[i][j]**((ρ-1)/ρ))**(ρ/(ρ-1)) - kf*result1[i][j]-knf*result2[i][j] - ω*((1-γ)*grid[i]*(i0*result1[i][j]**2/N))

    # Plot the results using a heatmap



    import seaborn as sns

    space = np.around(np.linspace(0,0.9,10),decimals=1)  # Determine the values of the axis
    
    size = np.around(np.linspace(0+int(n/20),n-int(n/20),10),decimals=0) 

    fig, ax = plt.subplots()
    sns.heatmap(result1,cbar_kws={"label":"$H_f$ value"},xticklabels =space,yticklabels=space)
    plt.title("$H_f$",fontsize=20)
    ax.set_xticks(size)
    ax.set_yticks(size)
    plt.xlabel("$c(TW)$")
    plt.ylabel("$β(HC)$")

    fig, ax = plt.subplots()
    sns.heatmap(result2,cbar_kws={"label":"$H_{nf}$ value"},xticklabels =space,yticklabels=space)
    plt.title("$H_{nf}$",fontsize=20)
    ax.set_xticks(size)
    ax.set_yticks(size)
    plt.xlabel("$c(TW)$")
    plt.ylabel("$β(HC)$")

    fig, ax = plt.subplots()
    sns.heatmap(H,cbar_kws={"label":"$H$ value"},xticklabels =space,yticklabels=space,vmin=0,vmax=1)
    plt.title("$H$",fontsize=20)
    ax.set_xticks(size)
    ax.set_yticks(size)
    plt.xlabel("$c(TW)$")
    plt.ylabel("$β(HC)$")

    fig, ax = plt.subplots()
    sns.heatmap(Hf_H,cbar_kws={"label":"$H_f/H$ value"},xticklabels =space,yticklabels=space)
    plt.title("$H_f/H$",fontsize=20)
    ax.set_xticks(size)
    ax.set_yticks(size)
    plt.xlabel("$c(TW)$")
    plt.ylabel("$β(HC)$")

    fig, ax = plt.subplots()
    sns.heatmap(I,cbar_kws={"label":"$I$ value"},xticklabels =space,yticklabels=space)
    plt.title("$Infections$",fontsize=20)
    ax.set_xticks(size)
    ax.set_yticks(size)
    plt.xlabel("$c(TW)$")
    plt.ylabel("$β(HC)$")

    fig, ax = plt.subplots()
    sns.heatmap(D,cbar_kws={"label":"Deads value"},xticklabels =space,yticklabels=space)
    plt.title("Deads",fontsize=20)
    ax.set_xticks(size)
    ax.set_yticks(size)
    plt.xlabel("$c(TW)$")
    plt.ylabel("$β(HC)$")
    
    fig, ax = plt.subplots()
    sns.heatmap(W,cbar_kws={"label":"Welfare value"},xticklabels =space,yticklabels=space)
    plt.title("Welfare",fontsize=20)
    ax.set_xticks(size)
    ax.set_yticks(size)
    plt.xlabel("$c(TW)$")
    plt.ylabel("$β(HC)$")
    
    fig, ax = plt.subplots()
    sns.heatmap(y_opt,cbar_kws={"label":"Output value"},xticklabels =space,yticklabels=space)
    plt.title("Output",fontsize=20)
    ax.set_xticks(size)
    ax.set_yticks(size)
    plt.xlabel("$c(TW)$")
    plt.ylabel("$β(HC)$")
    
    plt.show()
    
heat(20,1.1,100) # Show the results for Omega=20 , Rho = 1.1 and discretize the space in 100 points.
```
